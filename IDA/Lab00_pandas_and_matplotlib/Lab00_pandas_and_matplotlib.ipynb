<b>Exercise 5:</b>
* Display the mean churn rate of the clients.
* Now we want to analyse statistical information only for the clients, which are or aren't loyal to their telecom-company (field 'Churn' in dataframe). Extract the loyal and non-loyal clients from the table separately and display the means of their charateristics in a single dataframe.
* How long do the non-loyal users talk during the day (on average)?
* What is the maximum length of the international calls for the loyal users that do not use the international plan?

```python
### Write Your code here ###

# Calculate the mean churn rate
mean_churn_rate = tc_d['Churn'].mean()

# Extract loyal and non-loyal clients separately
loyal_clients = tc_d[tc_d['Churn'] == 0]
non_loyal_clients = tc_d[tc_d['Churn'] == 1]

# Display the means of their characteristics in a single DataFrame
loyal_mean = loyal_clients.mean(numeric_only=True)
non_loyal_mean = non_loyal_clients.mean(numeric_only=True)
loyalty_comparison = pd.DataFrame({'Loyal': loyal_mean, 'Non-Loyal': non_loyal_mean})

# Calculate the average length of daytime calls for non-loyal users
avg_day_calls_non_loyal = non_loyal_clients['Total day minutes'].mean()

# Find the maximum length of international calls for loyal users who do not use the international plan
max_intl_calls_loyal_non_intl_plan = loyal_clients[loyal_clients['International plan'] == 'No']['Total intl minutes'].max()

# Displaying the results
print("Mean churn rate of the clients:", mean_churn_rate)
print("\nMeans of characteristics for loyal and non-loyal clients:\n", loyalty_comparison)
print("\nAverage length of daytime calls for non-loyal users:", avg_day_calls_non_loyal)
print("\nMaximum length of international calls for loyal users without international plan:", max_intl_calls_loyal_non_intl_plan)
```

```python
### Write Your code here ###

# Build the cross table between the features using the method crosstab
cross_tab = pd.crosstab(index=tc_d['International plan'], columns=tc_d['Churn'])

# Visualize the distribution for the feature 'Churn' depending on the value of the features
fig, axes = plt.subplots(nrows=1, ncols=3, figsize=(15, 5))

# Distribution based on 'International plan'
sns.countplot(x='Churn', hue='International plan', data=tc_d, ax=axes[0])
axes[0].set_title('Churn distribution based on International plan')

# Distribution based on 'Voice mail plan'
sns.countplot(x='Churn', hue='Voice mail plan', data=tc_d, ax=axes[1])
axes[1].set_title('Churn distribution based on Voice mail plan')

# Distribution based on 'Customer service calls'
sns.countplot(x='Churn', hue='Customer service calls', data=tc_d, ax=axes[2])
axes[2].set_title('Churn distribution based on Customer service calls')

plt.tight_layout()
plt.show()
```
